```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
# plt.rcParams.update({
#     "text.usetex": True
# })
```

```python
scan_m4l = pd.read_csv('data/nll/nll-m4l.csv')
scan_unbinned = pd.read_csv('data/nll/nll-full_2.csv')
```

```python
plt.plot(scan_m4l['c6'], scan_m4l['nll'], label='m4l')
plt.plot(scan_unbinned['c6'], scan_unbinned['nll'], label='matrix-element')

plt.plot([-20.0,20.0],[1,1],'--',color='grey')
plt.text(-19.0,1.1,'$1\\sigma$')
plt.plot([-20.0,20.0],[4,4],'--',color='grey')
plt.text(-19.0,4.1,'$2\\sigma$')

plt.text(-9, 9.0, '$gg (\\to h^\\ast) \\to ZZ$', size='medium')
plt.text(-9, 8.5, '$ \\sqrt{s} = 14\\,\\mathrm{TeV},\\, 3\\,\\mathrm{ab}^{-1} $ (HL-LHC expected)', size='medium')

plt.xlabel('$c_6$', loc='right')
plt.ylabel('$-2 \\mathrm{ln} (\\lambda)$', loc='top')

plt.xlim(-20,20)
plt.ylim(0.0,10.0)

plt.tight_layout()
plt.savefig('nll-m4l_vs_full.pdf')
```
